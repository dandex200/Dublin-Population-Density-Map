```python
df_small_areas = pd.read_csv('./Data/SAPS_2022_Small_Area_UR_171024.csv')
df_small_areas
```

## Identify Needed GUID's from GEOJSON File Used for boundaries

```python
gdf = gpd.read_file('./Data/Dublin_SA1.geojson')
gdf
```

```python
guids = gdf['SA_GUID_2022'].unique()
guids
```

## Create DF from Population Data CSV of needed GUIDS with Small Area Total Populations

```python
filtered_df = df_small_areas[df_small_areas['GUID'].isin(guids)]
filtered_df
```

```python
result_df = filtered_df[['GUID', 'T1_1AGETM', 'T1_1AGETF']]
result_df
```

```python
result_df['Sum'] = result_df['T1_1AGETM'] + result_df['T1_1AGETF']
df_GUID_TOTALS = result_df[['GUID', 'Sum']]
df_GUID_TOTALS
```

### Create GEOJSON of Only GUID, Summed SA Population, and Geometry

```python
gdf_minimal = gdf[['geometry', 'SA_GUID_2022']]
gdf_minimal
```

```python
gdf_merged = gdf_minimal.merge(
    df_GUID_TOTALS,
    left_on='SA_GUID_2022',  
    right_on='GUID',         
    how='left'               
)
gdf_merged = gdf_merged.rename(columns={'Sum': 'total_pop'})
gdf_merged = gdf_merged.drop(columns=['SA_GUID_2022'])
gdf_merged.to_file('Dublin_SA1_TOTALPOP.geojson', driver='GeoJSON')
gdf_merged
```
